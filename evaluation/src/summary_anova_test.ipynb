```python
from operator import concat
from scipy import stats
import json
import os
import pandas as pd
import matplotlib.pyplot as plt

# Read data
each_data = pd.read_csv('../csv_results/booksum_summaries/summary-evaluation-each-results.csv')
concat_data = pd.read_csv('../csv_results/booksum_summaries/summary-evaluation-concat-results.csv')
threshold_data = pd.read_csv('../csv_results/booksum_summaries/summary-evaluation-threshold-results.csv')

for k in range(10):
    each_ = each_data['k'] == k
    concat_ = concat_data['k'] == k
    threshold_ = threshold_data['k'] == k

    print('k =', k+1)
    statistic, pvalue = stats.f_oneway(each_data[each_]['ROUGE-1 F1'], concat_data[concat_]['ROUGE-1 F1'], threshold_data[threshold_]['ROUGE-1 F1'])
    print("ANOVA P Value: {} < 0.05 = {}".format(
            pvalue, str(pvalue < 0.05)))

    if pvalue < 0.05:
        # Run a paired t test between all of them.
        # print("Running a paired t test between all of them.")
        statistic, e_c_pvalue = stats.ttest_rel(each_data[each_]['ROUGE-1 F1'], concat_data[concat_]['ROUGE-1 F1'])
        print("T test P Value: {} < 0.05 = {}.".format(e_c_pvalue, str(e_c_pvalue < 0.05)))
        if e_c_pvalue < 0.05:
            print("There is a significant difference between each and concat.")
        else:
            print("There is no significant difference between each and concat.")
        statistic, e_t_pvalue = stats.ttest_rel(each_data[each_]['ROUGE-1 F1'], threshold_data[threshold_]['ROUGE-1 F1'])
        print("T test P Value: {} < 0.05 = {}.".format(e_t_pvalue, str(e_t_pvalue < 0.05)))
        if e_t_pvalue < 0.05:
            print("There is a significant difference between each and threshold.")
        else:
            print("There is no significant difference between each and threshold.")
        statistic, c_t_pvalue = stats.ttest_rel(concat_data[concat_]['ROUGE-1 F1'], threshold_data[threshold_]['ROUGE-1 F1'])
        print("T test P Value: {} < 0.05 = {}.".format(c_t_pvalue, str(c_t_pvalue < 0.05)))
        if c_t_pvalue < 0.05:
            print("There is a significant difference between concat and threshold.")
        else:
            print("There is no significant difference between concat and threshold.")


```
